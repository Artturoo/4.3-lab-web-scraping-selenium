####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
driver = webdriver.Chrome(PATH,options = opciones)
driver.get(url)
````

````python
tabla=driver.find_element(By.XPATH, '/html/body')

filas=tabla.find_elements(By.CLASS_NAME, 'normal')


data=[]


for f in filas:
    
    elemen=f.find_elements(By.TAG_NAME, 'td')
    
    tmp=[]
    
    for e in elemen:
        
        tmp.append(e.text)
        
    data.append(tmp)

    
data
````

````python
for e in range(len(data)):
    data[e]=data[e][3:]
````

````python
data
````

````python
fecha=[]
latitud=[]
longitud=[]
area=[]


for i in range(len(data)):
    fecha.append(data[i][0])
    latitud.append(data[i][1:2])
    longitud.append(data[i][3:4])
    area.append(data[i][8])
    
````

````python
data
````

````python
cabeceras=driver.find_element(By.XPATH, '//*[@id="haut_tableau"]')

cabeceras=[c.text for c in cabeceras.find_elements(By.CLASS_NAME, 'th2')]


cabeceras
````

````python
cabeceras.pop(0)
````

````python
cabeceras.pop(6)
````

````python
cabeceras
````

````python
df=pd.DataFrame(data, columns=cabeceras)

df.head()
````

#### Top 10 languages by number of native speakers stored in a Pandas Dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
driver = webdriver.Chrome(PATH,options = opciones)
driver.get(url)
````

````python
tabla=driver.find_element(By.XPATH, '//*[@id="mw-content-text"]/div[1]/table[1]/tbody')

filas=tabla.find_elements(By.TAG_NAME, 'tr')


data=[]


for f in filas:
    
    elemen=f.find_elements(By.TAG_NAME, 'td')
    
    tmp=[]
    
    for e in elemen:
        
        tmp.append(e.text)
        
    data.append(tmp)

    
data
````

````python
cabeceras=driver.find_element(By.XPATH, '//*[@id="mw-content-text"]/div[1]/table[1]/thead')

cabeceras=[c.text for c in cabeceras.find_elements(By.TAG_NAME, 'th')]

driver.quit()

cabeceras
````

````python
df=pd.DataFrame(data, columns=cabeceras)

df.head()
````

````python
df.drop(['Language family', 'Branch'], axis = 'columns', inplace=True)
````

````python
df.head()
````

````python
driver.quit()
````
